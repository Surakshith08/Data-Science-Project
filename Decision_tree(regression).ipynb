 **Aim:** To build and evaluate a Decision Tree Regressor model for predicting motor vehicle claim
 amounts using vehicle features and analyze its performance with relevant metrics and visualizations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv('/content/car_insurance_claim.csv')
data.head()
```

```python
X = data.drop(['OLDCLAIM'], axis=1)
y = data['OLDCLAIM']
```

```python
#Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=999
)
```

```python
#Build Decision Tree Regressor
# ------------------------------
dtr = DecisionTreeRegressor(
    criterion='squared_error',    # for regression
    min_samples_split=int(len(X_train) * 0.10),
    random_state=999
)
dtr.fit(X_train, y_train)
```

```python
# 6. Predictions
y_pred = dtr.predict(X_test)
y_pred
```

```python
# 7. Evaluation metrics
# ------------------------------
mse = mean_squared_error(y_test, y_pred)
```

```python
r2 = r2_score(y_test, y_pred)
```

```python
print("Mean Squared Error:", mse)
print("R² Score:", r2)
```

```python
plt.figure(figsize=(16,10))
plot_tree(dtr, filled=True, feature_names=X.columns)
plt.show()
```

```python
plt.scatter(y_test, y_pred, alpha=0.7, color="orange")
plt.xlabel("Actual Claim Amount")
plt.ylabel("Predicted Claim Amount")
plt.title("Decision Tree Regression: Actual vs Predicted")
plt.show()
```
