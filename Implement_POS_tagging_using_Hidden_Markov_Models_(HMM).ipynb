```python
# Sample Sentence
sentence = "The cat sat on the mat."

# Tokenize and tag
tokens = word_tokenize(sentence)
pos_tags = pos_tag(tokens)

# Print tagged tokens
print("POS Tags:", pos_tags)
```

```python
from nltk.tag import hmm

# Training data (word-tag pairs)
train_data = [
    [("The", "DT"), ("cat","NN"),("sat", "VBD"), ("mat", "NN")],
    [("A","DT"), ("dog", "NN"), ("ran", "VBD")]
]

# Train HMM
trainer = hmm.HiddenMarkovModelTrainer()
hmm_tagger = trainer.train_supervised(train_data)

# Test
test_sentence = "The cat sat mat"
tags = hmm_tagger.tag(word_tokenize(test_sentence))
print("HMM tags:", tags)
```

**Observations:**

1. The code starts by performing a simple, rule-based POS tagging on the sentence "The cat sat on the mat" using NLTK's pos_tag function. This function is pre-trained and provides accurate tags like DT (Determiner) for "The" and NN (Noun) for "cat".

1. The second part of the code demonstrates a statistical approach using a Hidden Markov Model (HMM). An HMM is trained on a small, predefined set of labeled sentences.

1. The HMM learns the probabilities of state transitions (e.g., the probability of a Determiner being followed by a Noun) and output emissions (e.g., the probability of the word "cat" being a Noun).

1. When a new sentence, "The cat sat mat," is provided, the trained HMM uses these probabilities to predict the most likely sequence of POS tags for the words.

1. The output shows that the HMM successfully tags the test sentence, assigning the same tags as the pre-trained pos_tag function, even without the word "on" from the original sentence. This highlights how a statistically trained model can generalize from limited training data.
