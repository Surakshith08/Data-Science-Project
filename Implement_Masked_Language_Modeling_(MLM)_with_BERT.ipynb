**Aim:** To Implement Masked Language Modeling (MLM) with BERT.

**Theory:**

*   Masked Language Modeling (MLM) is a task where the model predicts masked tokens in a sentence.
*   BERT is a transformer-based model pre-trained on MLM, making it suitable for this task.

```python
from transformers import BertTokenizer, BertForMaskedLM
import torch
# Load pre-trained BERT
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertForMaskedLM.from_pretrained("bert-base-uncased")
# Input with [MASK]
sentence = "BERT is a [MASK] model."
inputs = tokenizer(sentence, return_tensors="pt")
```

```python
# Predict
with torch.no_grad():
    outputs = model(**inputs)
    predictions = outputs.logits
```

```python
# Get predicted token
mask_token_index = torch.where(inputs["input_ids"] == tokenizer.mask_token_id)[1]
predicted_token_id = predictions[0, mask_token_index].argmax(axis=-1)
predicted_token = tokenizer.decode(predicted_token_id)
print("Predicted word:", predicted_token)
```



**Observation:**

*   The code successfully loads the BERT tokenizer and model.
*   The masked token `[MASK]` in the input sentence is replaced with a predicted word.

**Conclusion:**

*   BERT effectively performs Masked Language Modeling by predicting the most likely word for the masked token based on the surrounding context.
