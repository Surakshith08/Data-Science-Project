# Writing to a text file using built-in open() function

```python
with open('example_test.txt', 'w') as file:
    file.write("This is an example text.")
print("file created successfully")
```

# Reading a text file using built-in open() function

```python
with open('example_test.txt', 'r') as file:
    data = file.read()
    print(data)
```

# Writing muliple line

```python
lines = [
    "This is the first line.",
    "This is the second line.",
    "This is the third line.",
    "This is the fourth line.",
    "This is the fifth line."
]

with open('example_multi.txt', 'w') as file:
    for li in lines:
        file.write(li + "\n")
```

```python
with open('example_multi.txt', 'r') as file:
    lines = file.readlines()

for line in lines:
    print(line.strip())
```

# Writing the DataFrame to a TSV file

```python
import pandas as pd

# Creating a DataFrame with multiple lines
data = {
    'Column1': ['Row1Col1', 'Row2Col1', 'Row3Col1'],
    'Column2': ['Row1Col2', 'Row2Col2', 'Row3Col2'],
    'Column3': ['Row1Col3', 'Row2Col3', 'Row3Col3']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.to_csv('tab_example.tsv', sep='\t', index=False)
```

# Creating a dictionary with multiple lines

```python
import json


data = {
    "employees": [
        {"name": "John", "age": 30, "department": "Sales"},
        {"name": "Anna", "age": 25, "department": "Marketing"},
        {"name": "Peter", "age": 35, "department": "HR"}
    ]
}

# Writing the dictionary to a JSON file
with open('json_example.json', 'w') as file:
    json.dump(data, file, indent=4)
```

```python
json.dump() function is used to write the dictionary to the file in JSON format.
```

# Reading the JSON file

```python
import json


with open('json_example.json', 'r') as file:
    data = json.load(file)
    print(data)
```

```python
import json

# Reading the JSON file
with open('json_example.json', 'r') as file:
    data = json.load(file)

# Printing the data in a more readable format
print("Employees Information:")
for employee in data['employees']:
    print(f"Name: {employee['name']}, Age: {employee['age']}, Department: {employee['department']}")
```
