```python
import pandas as pd
from sqlalchemy import create_engine

# Step 1: Load Sample Data
customer_data = {
    'Name': [' john doe ', 'JANE SMITH', 'Alice JOHNSON '],
    'Email': ['JOHN.DOE@example.COM', 'JANE.SMITH@sample.COM', 'ALICE.johnson@sample.com']
}
df = pd.DataFrame(customer_data)

# Print original data
print("Original Data:")
print(df)

# Step 2: Clean and Transform Data
# Trim whitespace from names
df['Name'] = df['Name'].str.strip()

# Standardize case
df['Name'] = df['Name'].str.title()  # Converts names to Title Case
df['Email'] = df['Email'].str.lower()  # Converts emails to Lower Case

# Split names into first and last names
df[['First Name', 'Last Name']] = df['Name'].str.split(expand=True)

# Extract domain from email addresses
df['Domain'] = df['Email'].str.split('@').str[1]

# Print cleaned data
print("\nCleaned Data:")
print(df)

# Step 3: Load Data into the Database
# Create a SQLite database connection
engine = create_engine('sqlite:///customer_data.db')

# Load the DataFrame into the SQL database
df.to_sql('customers', con=engine, if_exists='replace', index=False)

# Confirm that the data has been loaded successfully
loaded_df = pd.read_sql('customers', con=engine)
print("\nLoaded Data from Database:")
print(loaded_df)
```
